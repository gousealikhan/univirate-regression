```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data=pd.read_csv('/content/book1.csv')
print(data)

x=np.array(data[['x1','x2']])
y=np.array(data[['y']])
print(x)
print(y)

x_mean=np.mean(x,axis=0)
x_std=np.std(x,axis=0)

print(x_mean)
print(x_std)

x=(x-x_mean)/x_std
print(x)

x_b=np.c_[np.ones((len(x),1)),x]
print(x_b)

theta=np.zeros((x_b.shape[1],1))
print(theta)

alpha=0.01
iterations=1000

for i in range(iterations):
  y_pred=x_b.dot(theta)
  error=y_pred-y

  gradient=(2/len(y))*x_b.T.dot(error)
  theta=theta-(alpha*gradient)

print("optimized weights:",theta)

print("Optimized coefficients (theta):", theta[1:])
print("Optimized intercept:", theta[0])

data_test=pd.read_csv('/content/book1.csv')
print(data_test)
x_test=np.array(data_test[['x1','x2']])
print(x_test)

x_test=(x_test-np.mean(x_test,axis=0))/np.std(x_test,axis=0)
print(x_test)

x_test_b=np.c_[np.ones((len(x_test),1)),x_test]
print(x_test_b)

y_predicted=x_test_b.dot(theta)
print("Predicted value for new input:", y_predicted)
```
